```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://careers.nike.com/principal-software-engineer-full-stack-itc/job/R-65703")
page_data=loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_data}

    ### INSTRUCTION:
    The scraped text is from a job posting on a careers page.

    Extract the job requirements in a way that helps match a candidate’s projects.
    Return the output in valid JSON with the following keys:

    - role: Job title
    - experience_level: Required experience (e.g. Intern, Fresher, 0-2 years, 2+ years)
    - must_have_skills: Core technical skills explicitly required
    - good_to_have_skills: Optional or preferred skills
    - tech_stack: Technologies, frameworks, tools mentioned
    - responsibilities: Key responsibilities or expectations
    - description: Short summary of the role (2–3 lines)

    Rules:
    - Do NOT add any explanation or text outside JSON
    - Use lists for skills and tech_stack
    - If a field is missing, return an empty list or empty string

    ### VALID JSON (NO PREAMBLE):
    """
)



chain_extract=prompt_extract | llm
res=chain_extract.invoke({"page_data":page_data})
res=res.content
print(res)
```

```python
from langchain_core.output_parsers import JsonOutputParser

parser = JsonOutputParser()
```

```python
job_des=parser.parse(res)
print(job_des)
print(type(job_des))
```

```python
import uuid
import chromadb

client = chromadb.PersistentClient("vectorstore")
collection = client.get_or_create_collection(name="portfolio")

if collection.count() == 0:
    for _, row in df.iterrows():
        collection.add(
            documents=row["project_explanation"],   # semantic meaning
            metadatas={
                "links": row["Links"],               # resume evidence
                "techstack": row["techstack"]        # structured skills
            },
            ids=[str(uuid.uuid4())]
        )
```

```python
# developing a query for semantic searching in vector db

def build_nl_query(job):
    role = job.get("role", "")
    skills = ", ".join(job.get("must_have_skills", [])[:4])
    responsibilities = ", ".join(job.get("responsibilities", [])[:2])
    tech = ", ".join(job.get("tech_stack", [])[:3])

    query = (
        f"{role} role focused on {skills}, "
        f"involving {responsibilities}, "
        f"using technologies such as {tech}."
    )

    return " ".join(query.split())


query=build_nl_query(job_des)
query

```

```python
# use this query to sematically search throgh vectordb
# query="an engineer who is ready to work with real time systems and work with web sockets"
results = collection.query(
    query_texts=[query],   #  semantic NL query
    n_results=3               # top-k relevant projects
)

results

```

```python
resume_projects=[m["links"] for m in results["metadatas"][0]]
resume_projects
```

```python
covered_skills = set()

for meta in results["metadatas"][0]:
    for skill in meta["techstack"].split(","):
        covered_skills.add(skill.strip())

covered_skills = list(covered_skills)
covered_skills


```

```python
skills_to_prepare = set(job_des["must_have_skills"]) - set(covered_skills)
skills_to_prepare = list(skills_to_prepare)
skills_to_prepare
```

```python
prompt_guidance = PromptTemplate.from_template(
    """
    You are a practical career advisor.

    Your task is to present clear, direct instructions to a student applying for a job.
    Do not motivate, encourage, or evaluate.
    Do not infer or deduce new information.
    Only organize and present the given data as instructions.

    PROJECTS TO INCLUDE IN RESUME (GitHub links):
    {resume_projects}

    SKILLS ALREADY COVERED THROUGH PROJECTS:
    {covered_skills}

    SKILLS / TOPICS TO PREPARE FOR THIS JOB:
    {skills_to_prepare}

    TASK:
    Generate instructions that clearly state:
    1. Which projects should be added to the resume.
    2. Which skills are already covered by these projects.
    3. Which skills or topics the student should prepare before interviews.

    Output requirements:
    - Be direct and concise.
    - Use short paragraphs or numbered points.
    - Do not add encouragement, opinions, or explanations.
    - Do not invent skills, projects, or topics.
    - Do not use markdown or headings.
    - Output plain text only.

    INSTRUCTIONS:
    """
)
```

```python
guidance = (prompt_guidance | llm).invoke({
    "skills_to_prepare": skills_to_prepare,
    "covered_skills": covered_skills,
    "resume_projects": resume_projects
}).content

print(guidance)
```
