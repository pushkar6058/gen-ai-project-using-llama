```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://careers.nike.com/principal-software-engineer-full-stack-itc/job/R-65703")
page_data=loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_data}

    ### INSTRUCTION:
    The scraped text is from a job posting on a careers page.

    Extract the job requirements in a way that helps match a candidate’s projects.
    Return the output in valid JSON with the following keys:

    - role: Job title
    - experience_level: Required experience (e.g. Intern, Fresher, 0-2 years, 2+ years)
    - must_have_skills: Core technical skills explicitly required
    - good_to_have_skills: Optional or preferred skills
    - tech_stack: Technologies, frameworks, tools mentioned
    - responsibilities: Key responsibilities or expectations
    - description: Short summary of the role (2–3 lines)

    Rules:
    - Do NOT add any explanation or text outside JSON
    - Use lists for skills and tech_stack
    - If a field is missing, return an empty list or empty string

    ### VALID JSON (NO PREAMBLE):
    """
)



chain_extract=prompt_extract | llm
res=chain_extract.invoke({"page_data":page_data})
res=res.content
print(res)
```

```python
from langchain_core.output_parsers import JsonOutputParser

parser = JsonOutputParser()
```

```python
json_res=parser.parse(res)
print(json_res)
print(type(json_res))
```

```python
prompt_guidance = PromptTemplate.from_template(
    """
    You are a practical career advisor.

    Your task is to give clear, direct instructions to a student applying for a job.
    Do not motivate, encourage, or evaluate. Do not use emotional language.
    Only state what actions the student should take.

    MUST-HAVE SKILLS FROM JOB DESCRIPTION:
    {must_have_skills}

    STUDENT'S CURRENT SKILL SET:
    {my_skills}

    PROJECT GITHUB LINKS (projects already completed by the student):
    {resume_projects}

    TASK:
    Based strictly on the information above, generate instructions that cover:

    1. Projects to include in the resume (use the provided GitHub links).
    2. Skills that are already covered through these projects.
    3. Skills required by the job that are not covered through projects.
    4. Topics the student should prepare or revise before interviews.

    Output requirements:
    - Be direct and concise.
    - Use short paragraphs or numbered points.
    - Do not add encouragement or opinions.
    - Do not invent skills or projects.
    - Do not use markdown or headings.
    - Output plain text only.

    INSTRUCTIONS:
    """
)
```

```python
guidance = (prompt_guidance | llm).invoke({
    "must_have_skills": must_have_skills,
    "my_skills": myskills,
    "resume_projects": resume_projects
}).content

print(guidance)
```
